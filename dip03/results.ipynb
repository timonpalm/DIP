```python
import pandas as pd
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
#freq_conv = pd.read_csv("benchmark_FM_FREQUENCY_CONVOLUTION.csv", header=1, delimiter=";", index_col=0)
#sep_filter = pd.read_csv("benchmark_FM_SEPERABLE_FILTER.csv", header=1, delimiter=";", index_col=0)
#spat_conv = pd.read_csv("benchmark_FM_SPATIAL_CONVOLUTION.csv", header=1, delimiter=";", index_col=0)

freq_conv = np.genfromtxt('benchmark_FM_FREQUENCY_CONVOLUTION.csv', delimiter=';', skip_header=1)
sep_filter = np.genfromtxt('benchmark_FM_SEPERABLE_FILTER.csv', delimiter=';', skip_header=1)
spat_conv = np.genfromtxt('benchmark_FM_SPATIAL_CONVOLUTION.csv', delimiter=';', skip_header=1)

filter_size = (freq_conv[0])[1:]
pic_size = (freq_conv[:,0])[1:]

print(filter_size)

freq_conv = freq_conv[1:,1:]
sep_filter = sep_filter[1:,1:]
spat_conv = spat_conv[1:,1:]

X,Y = np.meshgrid(filter_size, pic_size)
X.shape
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection='3d')
ax.set_title("FREQUENCY_CONVOLUTION")
ax.plot_surface(X, Y, freq_conv,alpha=1)
ax.view_init(30, -30)
ax.set_xlabel("kernel size")
ax.set_ylabel("image size")
ax.set_zlabel("time in s")
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection='3d')
ax.set_title("SEPERABLE_FILTER")
ax.plot_surface(X, Y, sep_filter, alpha=1)
ax.view_init(30, -30)
ax.set_xlabel("kernel size")
ax.set_ylabel("image size")
ax.set_zlabel("time in s")
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection='3d')
ax.set_title("SPATIAL_CONVOLUTION")
ax.plot_surface(X, Y, spat_conv,alpha=1)
ax.view_init(30, -90)
ax.set_xlabel("kernel size")
ax.set_ylabel("image size")
ax.set_zlabel("time in s")
```
